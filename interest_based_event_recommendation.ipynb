```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load datasets
#first_responses = pd.read_csv("first_responses.csv")
#events = pd.read_csv("events.csv")

# Drop rows with NaN values in 'All Interest' and 'Description' columns
first_responses.dropna(subset=['All Interests'], inplace=True)
events.dropna(subset=['Description'], inplace=True)

# Feature Engineering
tfidf = TfidfVectorizer()
interest_matrix = tfidf.fit_transform(first_responses["All Interests"].astype(str))
event_desc_matrix = tfidf.transform(events["Description"].astype(str))
similarity_matrix = cosine_similarity(interest_matrix, event_desc_matrix)

# Recommendation
# Recommendation
# Recommendation
def recommend_events(user_name, top_n=20, min_events_per_category=3):
    user_data = first_responses[first_responses["Name"] == user_name]
    if user_data.empty:
        return "User not found."

    user_interests = user_data["All Interests"].values[0].split(', ')
    user_interests_lower = [interest.lower() for interest in user_interests]

    user_index = first_responses[first_responses["Name"] == user_name].index[0]
    sim_scores = list(enumerate(similarity_matrix[user_index]))
    sim_scores.sort(key=lambda x: x[1], reverse=True)

    recommended_events = []
    category_count = {category: 0 for category in ["music and entertainment", "sports", "fitness", "dance", "adventure", "tech and innovation", "travel", "acting", "astronomy", "virtual events", "food and drinks", "books and education", "art and culture","photography"]}
    recommended_event_ids = set()

    for idx, _ in sim_scores:
        event_categories = events.iloc[idx]["Primary Category"].lower().split(', ')
        common_interests = set(event_categories) & set(user_interests_lower)

        if common_interests:  # If there are common interests between user and event
            for category in event_categories:
                category = category.strip()
                if category in category_count:
                    if category_count[category] < min_events_per_category:
                        event_id = events.iloc[idx]["Event ID"]
                        if event_id not in recommended_event_ids:  # Check if event is not recommended already
                            recommended_events.append(event_id)
                            recommended_event_ids.add(event_id)
                            category_count[category] += 1
                            break  # Break inner loop to avoid duplicate event IDs

    # If there are not enough events per category, recommend more events from categories with fewer recommendations
    for idx, _ in sim_scores:
        event_categories = events.iloc[idx]["Primary Category"].lower().split(', ')
        common_interests = set(event_categories) & set(user_interests_lower)

        if common_interests:  # If there are common interests between user and event
            for category in event_categories:
                category = category.strip()
                if category in category_count:
                    if category_count[category] < min_events_per_category:
                        event_id = events.iloc[idx]["Event ID"]
                        if event_id not in recommended_event_ids:  # Check if event is not recommended already
                            recommended_events.append(event_id)
                            recommended_event_ids.add(event_id)
                            category_count[category] += 1

    return recommended_events[:top_n]


# Example usage
user_name = "Shreyas Saboo"
recommended_events = recommend_events(user_name)
print(f"Recommended events for {user_name}:")
for event_id in recommended_events:
    print(f"Event ID: {event_id}")
```

### Code Flow and Explanation:

1. **Import Libraries**:
   - `pandas` for data manipulation.
   - `TfidfVectorizer` from `sklearn.feature_extraction.text` for converting text to numerical features.
   - `cosine_similarity` from `sklearn.metrics.pairwise` for measuring similarity between user interests and event descriptions.

2. **Load Data**:
   - Load user responses (`first_responses.csv`) and event details (`events.csv`) using `pandas`.
   - Drop rows where the 'All Interests' in user responses or the 'Description' in events are missing.

3. **Feature Engineering**:
   - Use **TF-IDF (Term Frequency-Inverse Document Frequency)** to vectorize the 'All Interests' of users and the 'Description' of events, creating numerical representations of these text fields.
   - Compute a **cosine similarity matrix** between the interest matrix and the event description matrix.

4. **Recommendation Function `recommend_events()`**:
   - Recommends events based on user interests and event descriptions using the similarity matrix.
   
   - **Steps**:
     - Retrieve the user's interests from the user responses dataset.
     - Match user interests to event categories by computing similarities.
     - Keep track of recommended events for each category, ensuring that at least a specified minimum number of events are recommended per category.
     - Avoid recommending the same event multiple times.
     - If there aren’t enough events per category, recommend additional events based on fewer common interests.
   
5. **Recommendation Logic**:
   - The function recommends events by checking for common interests between user interests and event categories, ensuring a minimum of `min_events_per_category` per category.
   - The top `n` recommendations (default is 20) are returned.

6. **Example Usage**:
   - Example demonstrates the recommendation process for a user named **Shreyas Saboo**.
   - Prints out the recommended event IDs for the user.
